```python
cols = list(df.columns)

new_df = df[cols[0:3]]

new_df
```

```python
new_df.sort_values('Ladder score', ascending = False)
```

```python
new_df = new_df.rename(columns = {'Ladder score':'Happiness Score'})
new_df
```

```python
regions = list(df['Regional indicator'])
regions

regions_single = list(set([x for x in regions if regions.count(x)>1]))
regions_single
```

```python
new_df.loc[df['Regional indicator'] == 'Western Europe']
```

```python
frames = []

for r in regions_single:
    reg_df = new_df.loc[df['Regional indicator'] == r]
    frames.append(reg_df)
    
final_df = pd.concat(frames)

final_df.reset_index(drop=True, inplace=True)

score_by_region = final_df
```
